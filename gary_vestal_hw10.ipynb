```python
st = stopwords.words('english')
stemmer = PorterStemmer()

def loadDataAsDataFrame(f_path):
    '''
        Given a path, loads a data set and puts it into a dataframe
        - simplified mechanism
    '''
    df = pd.read_csv(f_path)
    return df


def preprocess_text(raw_text):

    # Replace/remove username
    raw_text = re.sub('(@[A-Za-z0-9\_]+)', 'atUsername', raw_text)
    #replace website names
    #from https://stackoverflow.com/questions/42618872/regex-for-website-or-url-validation/42619368
    raw_text = re.sub('(http[A-Za-z0-9\_]+)', 'toWebsite', raw_text)
    #replace non-alpha-numeric with space
    raw_text = re.sub('\W', ' ', raw_text)
    
    #stemming and lowercasing
    words=[]
    for w in raw_text.lower().split():
        if not w in st and not w in ['.',',', '[', ']', '(', ')', ':']:
            words.append(w)
            
    return (" ".join(words))

from nltk import word_tokenize
```

```python
from gensim.models import Word2Vec
from gensim.models import KeyedVectors

filename = 'DSM-language-model-1B-LARGE/DSM-language-model-1B-LARGE/trig-vectors-phrase.bin'
print ('loading model this may take a while')
model = KeyedVectors.load_word2vec_format(filename, binary=True, encoding='utf8',unicode_errors='ignore')

print (model.most_similar("paxil"))
print (model.most_similar("adderall"))
print (model.most_similar("headache"))
```

```python
drug="tamoxifen"

model.most_similar(drug, topn=100)
```

```python
import Levenshtein

ratios={}

for (word, score) in model.most_similar(drug, topn=10000):
    #print(word, score)
    similarity_score = Levenshtein.ratio(word, drug)
    #print(similarity_score)
    ratios[word]=similarity_score
```

```python
for thresh in [0.8, 0.74, 0.7, 0.6]:
    print("Threshold: ",thresh)
    print("Spelling\tLev Ratio")
    for word, ratio in ratios.items():
        if ratio >= thresh:
            print(word, ratio)
    print("*****************")
```
